```python
import pandas as pd
import numpy as np

df = pd.read_csv("Traffic_Crashes_-_Crashes.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Checking the percentage of missing values in each column
missing_data = df.isnull().sum() / len(df) * 100

# Display columns with more than 0% missing values
s=missing_data[missing_data > 0].sort_values(ascending=False)
print(s)
```

```python
# Dropping columns with high missing data

high_missing = ['WORKERS_PRESENT_I','DOORING_I','WORK_ZONE_TYPE','WORK_ZONE_I','PHOTOS_TAKEN_I','STATEMENTS_TAKEN_I','NOT_RIGHT_OF_WAY_I','CRASH_DATE_EST_I','INTERSECTION_RELATED_I', 'LANE_CNT', 'HIT_AND_RUN_I']
df = df.drop(columns=high_missing)

df.head()
```

```python
# imputing the missing vaLues in longitude and latitude with numeric values

median_longitude = df['LONGITUDE'].median()
median_latitude = df['LATITUDE'].median()
df['LONGITUDE'].fillna(median_longitude, inplace=True)
df['LATITUDE'].fillna(median_latitude, inplace=True)

df.head()
```

```python
# For categorical columns, we use the mode or a placeholder like 'Unknown'

for column in ['REPORT_TYPE', 'LOCATION', 'STREET_DIRECTION', 'STREET_NAME', 'RD_NO', 'MOST_SEVERE_INJURY']:
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)
    
df.head()
```

```python
# For injury-related columns, we assume missing values mean no injuries, so we fill with 0

injury_columns = ['INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING',
                  'INJURIES_REPORTED_NOT_EVIDENT', 'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN', 'INJURIES_TOTAL']
df[injury_columns] = df[injury_columns].fillna(0)

df.head()
```

```python


# Checking the remaining missing values if any
remaining_missing_cleaning = df.isnull().sum() / len(df) * 100
remaining_missing_cleaning[remaining_missing_cleaning > 0]
print(remaining_missing_cleaning)
```

```python

# Imputing missing values in 'BEAT_OF_OCCURRENCE' with the mode

beat_of_occurrence = df['BEAT_OF_OCCURRENCE'].mode()[0]
df['BEAT_OF_OCCURRENCE'].fillna(beat_of_occurrence, inplace=True)
df.head()
```

```python
# Final check for any remaining missing values
final_missing_check = df.isnull().sum() / len(df) * 100
final_missing_check[final_missing_check > 0]

print(final_missing_check)
```
